```python
def create_DFA(reg):
    nfa=regex_to_nfa.regex_to_nfa(reg)
    dfa=nfa_to_dfa.nfa_to_dfa(nfa)
    dfa=dfa_to_efficient_dfa.dfa_to_efficient_dfa(dfa)
    visual_utils.draw_dfa(dfa, title="")
```
